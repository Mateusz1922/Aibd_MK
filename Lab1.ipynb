```python
import matplotlib.pyplot as plt
import numpy as np
```

Zad 3.

```python
f = lambda x: x**2+5

x1 = np.linspace(-1,1,100)
y1 = f(x1)
x2 = np.linspace(-6,6,100)
y2 = f(x2)
x3 = np.linspace(0,5,100)
y3 = f(x3)

plt.plot(x1,y1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Wykres funkcji f(x)=x^2+5, x należy do przedziału (-1,1)')
plt.legend('x')
plt.show()

plt.plot(x2,y2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Wykres funkcji f(x)=x^2+5, x należy do przedziału (-6,6)')
plt.legend('x')
plt.show()

plt.plot(x3,y3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Wykres funkcji f(x)=x^2+5, x należy do przedziału (0,5)')
plt.legend('x')
plt.show()
```

W zadaniu należało zdefiniować zadaną funkcję, a następnie sporządzić jej wykres dla zadanych argumentów. Funkcję zdefiniowałem metodą lambda, a wektor argumentów stworzyłem za pomocą funkcji np.linspace. Przy użyciu narzędzi matplotlib narysowałem wykresy funkcji razem z odpowiednio dodanymi elementami wykresów.
