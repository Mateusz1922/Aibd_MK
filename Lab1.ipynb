```python
import matplotlib.pyplot as plt
import numpy as np
```

Zad 3.

```python
f = lambda x: x**2+5

x1 = np.linspace(-1,1,100)
y1 = f(x1)
x2 = np.linspace(-6,6,100)
y2 = f(x2)
x3 = np.linspace(0,5,100)
y3 = f(x3)

plt.plot(x1,y1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Wykres funkcji f(x)=x^2+5, x naley do przedziau (-1,1)')
plt.legend('x')
plt.show()

plt.plot(x2,y2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Wykres funkcji f(x)=x^2+5, x naley do przedziau (-6,6)')
plt.legend('x')
plt.show()

plt.plot(x3,y3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Wykres funkcji f(x)=x^2+5, x naley do przedziau (0,5)')
plt.legend('x')
plt.show()
```

W zadaniu naleao zdefiniowa zadan funkcj, a nastpnie sporzdzi jej wykres dla zadanych argumentw. Funkcj zdefiniowaem metod lambda, a wektor argumentw stworzyem za pomoc funkcji np.linspace. Przy uyciu narzdzi matplotlib narysowaem wykresy funkcji razem z odpowiednio dodanymi elementami wykresw.
